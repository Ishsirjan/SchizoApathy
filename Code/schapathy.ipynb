```python
# SchizApathy Mini-Project: Functional Connectivity in Apathy-Linked Regions

# --- SETUP ---
import os
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
import gzip
from nilearn import datasets, image, plotting, input_data, connectome
import nibabel as nib

# --- PARAMETERS ---
with open("/home/cgntlab4/Downloads/4197885/downloaded_subjects.txt") as f:
    downloaded_subjects = [line.strip() for line in f.readlines() if line.strip()]

subject_ids = downloaded_subjects

# Group assignment for demo purposes (should be replaced with real mapping)
import gzip

# Load phenotypic file to get real group labels
pheno_df = pd.read_csv(os.path.join(base_dir, 'phenotypic_data.tsv'), sep='	')
pheno_df['StrID'] = pheno_df['ID'].astype(str).str.zfill(7)
pheno_df = pheno_df[pheno_df['StrID'].isin(subject_ids)]
sz_subjects = set(pheno_df[pheno_df['Subject Type'] == 'Patient']['StrID'])
hc_subjects = set(pheno_df[pheno_df['Subject Type'] == 'Control']['StrID'])

base_dir = '/home/cgntlab4/Downloads/4197885'

# --- LOAD A DIFFERENT ATLAS (DK instead of Destrieux) ---
atlas = datasets.fetch_atlas_schaefer_2018(n_rois=100, yeo_networks=7, resolution_mm=2)
atlas_img = atlas['maps']
atlas_labels = [label.decode('utf-8') if isinstance(label, bytes) else label for label in atlas['labels']]

# Use all ROIs for now
roi_indices = list(range(len(atlas_labels)))
selected_regions = atlas_labels
print("Selected brain regions from Schaefer 100 ROI atlas:")
for region in selected_regions:
    print("-", region)

# --- FUNCTION: Load confounds safely ---
def load_confounds(confound_path):
    try:
        return pd.read_csv(confound_path, sep='\t')
    except Exception as e:
        print(f"Error reading confounds for {confound_path}: {e}")
        return None

# --- FUNCTION: Process subject ---
def process_subject(sub_id):
    bold_path = os.path.join(base_dir, f'fmri_{sub_id}.nii.gz')
    confound_path = os.path.join(base_dir, f'fmri_{sub_id}.tsv.gz')

    if not os.path.exists(bold_path) or not os.path.exists(confound_path):
        print(f"Missing files for subject {sub_id}")
        return None

    confounds_df = load_confounds(confound_path)
    if confounds_df is None:
        return None

    confounds_numeric = confounds_df.select_dtypes(include=[np.number]).fillna(0)

    subject_img = nib.load(bold_path)
    resampled_atlas = image.resample_to_img(atlas_img, subject_img, interpolation='nearest')

    masker = input_data.NiftiLabelsMasker(
        labels_img=resampled_atlas,
        standardize=True,
        memory='nilearn_cache')

    time_series = masker.fit_transform(subject_img, confounds=confounds_numeric)
    return time_series

# --- VISUAL CHECK: PLOT FUSED IMAGE ---
from nilearn.plotting import plot_roi

example_subject = subject_ids[0]
example_img_path = os.path.join(base_dir, f'fmri_{example_subject}.nii.gz')

if os.path.exists(example_img_path):
    example_img = nib.load(example_img_path)
    resampled = image.resample_to_img(atlas_img, example_img, interpolation='nearest')
    plot_roi(resampled, bg_img=image.index_img(example_img, 0), title=f"Atlas on subject {example_subject}", display_mode='ortho', cut_coords=(0, 0, 0), output_file="atlas_overlay.png")
    print("Saved atlas overlay as atlas_overlay.png")
else:
    print(f"Example subject not found: {example_img_path}")

# --- EXTRACT TIME SERIES ---
subject_data = {}
for sub in subject_ids:
    data = process_subject(sub)
    if data is not None:
        print(f"✔ Processed subject {sub}: shape = {data.shape}")
        subject_data[sub] = data
    else:
        print(f"✘ Failed subject {sub}")

# --- COMPUTE CONNECTIVITY MATRICES ---
def compute_fc_matrix(timeseries):
    return np.corrcoef(timeseries.T) if timeseries.size > 0 else None

connectivity_matrices = {
    sub: compute_fc_matrix(ts) for sub, ts in subject_data.items() if compute_fc_matrix(ts) is not None
}

# --- GROUP ANALYSIS WITH STATS ---
from scipy.stats import ttest_ind

sz_matrices = [mat for sid, mat in connectivity_matrices.items() if sid in sz_subjects]
hc_matrices = [mat for sid, mat in connectivity_matrices.items() if sid in hc_subjects]

sz_avg = np.mean(sz_matrices, axis=0) if sz_matrices else np.array([])
hc_avg = np.mean(hc_matrices, axis=0) if hc_matrices else np.array([])
diff_matrix = sz_avg - hc_avg if sz_avg.size and hc_avg.size else np.array([])

# --- STATISTICAL TESTING ---
from itertools import combinations
if sz_matrices and hc_matrices:
    sz_array = np.stack(sz_matrices)
    hc_array = np.stack(hc_matrices)
    t_stats, p_vals = ttest_ind(sz_array, hc_array, axis=0)
    
    # Exclude diagonals from stats
    np.fill_diagonal(t_stats, 0)
    np.fill_diagonal(p_vals, 1)

    # Save top differences
    np.savetxt("t_stat_matrix.csv", t_stats, delimiter=",", header=",".join(atlas_labels), comments='')
    np.savetxt("p_val_matrix.csv", p_vals, delimiter=",", header=",".join(atlas_labels), comments='')

    # Plot significance map (thresholded for p < 0.05)
    sig_mask = p_vals < 0.05
    diff_sig = diff_matrix * sig_mask

    plt.figure(figsize=(10, 8))
    sns.heatmap(diff_sig, cmap='coolwarm', center=0, square=True)
    plt.title("Significant Differences (p < 0.05)")
    plt.tight_layout()
    plt.savefig("fc_significant_diff.png")
    plt.close()

    print("Saved statistical comparison plots and CSVs.")

# --- BAR PLOT SUMMARY: WITHIN VS BETWEEN NETWORKS ---
    def get_network(label):
        label = label.decode('utf-8') if isinstance(label, bytes) else label
        return label.split('_')[2]

    network_conn = {}
    for i, j in combinations(range(len(atlas_labels)), 2):
        net_i = get_network(atlas_labels[i])
        net_j = get_network(atlas_labels[j])
        key = f"{net_i}-{net_j}" if net_i != net_j else f"{net_i}-{net_i}"
        if key not in network_conn:
            network_conn[key] = []
        network_conn[key].append(diff_matrix[i, j])

    avg_conn = {k: np.mean(v) for k, v in network_conn.items()}
    sorted_items = sorted(avg_conn.items(), key=lambda x: x[1], reverse=True)

    plt.figure(figsize=(12, 6))
    sns.barplot(x=[k for k, _ in sorted_items], y=[v for _, v in sorted_items])
    plt.xticks(rotation=90)
    plt.ylabel("Mean SZ-HC FC Difference")
    plt.title("Mean Functional Connectivity Differences by Network Pair")
    plt.tight_layout()
    plt.savefig("network_barplot.png")
    plt.close()
    print("Saved bar plot of within/between network differences.")

# --- APATHY-FOCUSED ANALYSIS ---

# Define apathy-related keywords
apathy_keywords = ['cingulate', 'frontal', 'orbitofrontal', 'insula', 'amygdala', 'OFC', 'PFC']

# Find indices of ROIs that match apathy regions
apathy_indices = [i for i, label in enumerate(atlas_labels) if any(k.lower() in label.lower() for k in apathy_keywords)]

# Print out apathy-relevant ROI pairs with strongest differences
print(">>> TOP APATHY-RELATED FC DIFFERENCES <<<")
if sz_matrices and hc_matrices and len(apathy_indices) > 0:
    apathy_pairs = []
    for i in apathy_indices:
        for j in range(len(atlas_labels)):
            if i != j:
                apathy_pairs.append((i, j, t_stats[i, j], p_vals[i, j]))

    top_apathy = sorted(apathy_pairs, key=lambda x: abs(x[2]), reverse=True)[:10]
    # Export to CSV
    apathy_df = pd.DataFrame([
        {'ROI 1': atlas_labels[i], 'ROI 2': atlas_labels[j], 't-stat': t, 'p-val': p}
        for i, j, t, p in top_apathy
    ])
    apathy_df.to_csv("apathy_top_differences.csv", index=False)

    for i, j, t, p in top_apathy:
        print(f"{atlas_labels[i]} ↔ {atlas_labels[j]} | t = {t:.3f}, p = {p:.4f}")
else:
    print("No apathy-related ROI matches found or insufficient data.")

# --- TOP REGION PAIRS AND NETWORK SUMMARY ---

# Convert atlas_labels from bytes to strings
atlas_labels = [label.decode('utf-8') if isinstance(label, bytes) else label for label in atlas_labels]

if sz_matrices and hc_matrices:
    # Print top 10 region pairs with highest absolute t-stats
    top_n = 10
    flat_indices = np.argsort(np.abs(t_stats).ravel())[::-1][:top_n]
    rows, cols = np.unravel_index(flat_indices, t_stats.shape)
    print("--- Top region pairs by absolute t-statistic ---")
    for i in range(top_n):
        r1, r2 = rows[i], cols[i]
        print(f"{atlas_labels[r1]} ↔ {atlas_labels[r2]} | t = {t_stats[r1, r2]:.3f}, p = {p_vals[r1, r2]:.4f}")

    # Summary: count top 50% different edges by network name root
    network_counts = {}
    for i in range(len(atlas_labels)):
        root = atlas_labels[i].split('_')[2]  # 7Networks_LH_Default_X → Default
        network_counts[root] = network_counts.get(root, 0) + 1

    print("--- ROI count by network group ---")
    for net, count in sorted(network_counts.items(), key=lambda x: -x[1]):
        print(f"{net}: {count} ROIs")

# --- VISUALIZE APATHY CONNECTIONS ---
    from nilearn.plotting import plot_connectome
    from nilearn import plotting as nlp

    # Extract mean coordinates of atlas regions
    example_img = nib.load(example_img_path)
    resampled_atlas_for_plot = image.resample_to_img(atlas_img, example_img, interpolation='nearest')
    coords = plotting.find_parcellation_cut_coords(labels_img=resampled_atlas_for_plot)

    apathy_conn = np.zeros((len(atlas_labels), len(atlas_labels)))
    for i, j, _, _ in top_apathy:
        apathy_conn[i, j] = diff_matrix[i, j]
        apathy_conn[j, i] = diff_matrix[i, j]  # symmetric

    plot_connectome(apathy_conn, coords, node_size=20,
                    title="Top Apathy-Related FC Differences",
                    output_file="apathy_connectome.png")
    print("Saved apathy-related connectome visualization.")

# --- VISUALIZATION ---
def plot_and_save(matrix, title, filename):
    if matrix.size == 0:
        print(f"Skipping plot for {title}: empty matrix")
        return
    plt.figure(figsize=(10, 8))
    sns.heatmap(matrix, cmap='coolwarm', center=0, square=True)
    plt.title(title)
    plt.tight_layout()
    plt.savefig(filename)
    plt.close()

plot_and_save(sz_avg, "Average FC - Schizophrenia", "fc_schizophrenia.png")
plot_and_save(hc_avg, "Average FC - Control", "fc_control.png")
plot_and_save(diff_matrix, "Difference FC (SZ - HC)", "fc_difference.png")

print("Saved group-level connectivity plots as PNGs.")
```
